# Uso Librería para Brazos Roboticos con 3 Grados de Libertad
## Ejemplo uso de trayectorias bresenham2D con Cinemática Inversa
Cargamos las librerias que vamos a usar:
- **Brazobot3GL** es la librería que queremos probar
- **miUtilidades** es una librería en las que incluyo eso, métodos y utilidades para varias cosas

```python
from Brazobot3GL import Brazobot3GL
from serial import Serial
from ipywidgets import interact
import misUtilidades
import math
```

Obtenemos la conexión a nuestro Arduino
- Esta librería se ha hecho para que le pasemos la conexión al puerto serie donde esté nuestro arduino conectado, por lo que debemos tener uno.
- Con la función **misUtilidades.getSPArduino()** obtenemos automáticamente el puerto donde está conectado el arduino. Si, si, automáticamente, sin mirar puertos ni nada y está probado en MAC y Windows y con arduino UNO y Nano. 
    
**¡¡¡Completamente Plug&Play!!!**

```python
puertoSP = misUtilidades.getSPArduino()
sp = Serial(puertoSP, 19200)
```

Creamos nuestro objeto <BrazoRobotico>. En el constructor le pasamos el puerto serie del arduino, la altura a la que se encuentra la primera articulación, Si hay cabeceo, y el angulo del cabeceo

```python
RA3GL = Brazobot3GL(sp, 514, 'Y', 0)
```

**Definimos las articulaciones** del Brazo Robotico

```python
RA3GL.setBrazo([0, 160],546)
RA3GL.setAnteBrazo([20, 180],854)
RA3GL.setMuneca([0, 180],455)
```

**Posicionamos el brazo robotico** en una posición cualquiera mediante Cinemática Inversa (X, Y)

```python
RA3GL.posIK(455, 1914)
```

Movemos el brazo a otra posición mediante **trayectorias de Bresenham de 2D** e IK

```python
RA3GL.posXYTray(882, 0)
```
